```
import pickle

nLabel, nTrial, nUser, nChannel, nTime  = 4, 40, 32, 40, 8064
no_of_users=32
def convertData():
    print("Program started"+"\n")
    fout_data = open("/content/drive/My Drive/data/features_raw.dat",'w')
    fout_labels0 = open("/content/drive/My Drive/data/labels_0.dat",'w')
    fout_labels1 = open("/content/drive/My Drive/data/labels_1.dat",'w')
    fout_labels2 = open("/content/drive/My Drive/data/labels_2.dat",'w')
    fout_labels3 = open("/content/drive/My Drive/data/labels_3.dat",'w')
    for i in range(no_of_users):  #nUser #4, 40, 32, 40, 8064 4 labels, 40 sample for each user, 32 such user, 40 electrode, 8064*40 features
        if(i%1 == 0):
            if i < 10:
                name = '%0*d' % (2,i+1)
            else:
                name = i+1
        fname = "/content/drive/My Drive/data_preprocessed_python/s"+str(name)+".dat"     
        f = open(fname, 'rb')                 #Read the file in Binary mode
        x = pickle.load(f, encoding='latin1')
        print(fname)                          
    	
        for tr in range(nTrial):
            if(tr%1 == 0):
                for dat in range(nTime):
                    if(dat%32 == 0):
                        for ch in range(nChannel):
                            fout_data.write(str(x['data'][tr][ch][dat]) + " ");
                fout_labels0.write(str(x['labels'][tr][0]) + "\n");
                fout_labels1.write(str(x['labels'][tr][1]) + "\n");
                fout_labels2.write(str(x['labels'][tr][2]) + "\n");
                fout_labels3.write(str(x['labels'][tr][3]) + "\n");
                fout_data.write("\n");
    fout_labels0.close()
    fout_labels1.close()
    fout_labels2.close()
    fout_labels3.close()
    fout_data.close()
    print("\n"+"Print Successful")
```

```
def sampleFeatures():
    print("Program started"+"\n")
    fout_data = open('/content/drive/My Drive/data/features_sampled.dat','w')
    for i in range(no_of_users):   #nUser  #4, 40, 32, 40, 8064
        if(i%1 == 0):
            if i < 10:			
                name = '%0*d' % (2,i+1)
            else:
                name = i+1		
            fname = "/content/drive/My Drive/data_preprocessed_python/s"+str(name)+".dat"
            x = pickle.load(open(fname, 'rb'), encoding='latin1')		
            print(fname)		
            for tr in range(nTrial):			
                if(tr%1 == 0):				
                    for dat in range(nTime):					
                        if(dat%32 == 0 ):						
                            for ch in range(nChannel):							
                                fout_data.write(str(ch+1) + " ");							
                                fout_data.write(str(x['data'][tr][ch][dat]) + " ");			
                fout_data.write("\n");
    fout_data.close()
    print('Completed')
```

```
import numpy
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

def svm_classifier(): 
    file_x = 'data/features_sampled.dat'
    file_y = 'data/label_class_0.dat'
    
    X = numpy.genfromtxt(file_x, delimiter=' ')
    y = numpy.genfromtxt(file_y, delimiter=' ')
    
    # Split the data into training/testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
    
    # Feature Scaling
    from sklearn.preprocessing import StandardScaler
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    	
    # SVM Classifier
    clf = SVC(kernel = 'rbf', random_state = 42)
    clf.fit(X_train, y_train)
    y_predict = clf.predict(X_test)
    cm = confusion_matrix(y_test, y_predict)
    print(cm)
    print("Accuracy score of Valence ")
    print(accuracy_score(y_test, y_predict)*100)
    
    ###############################################################
    
    file_x = 'data/features_sampled.dat'
    file_y = 'data/label_class_1.dat'
    
    X = numpy.genfromtxt(file_x, delimiter=' ')
    y = numpy.genfromtxt(file_y, delimiter=' ')
    
    # Split the data into training/testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
    
    # Feature Scaling
    from sklearn.preprocessing import StandardScaler
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    	
    # SVM Classifier
    clf = SVC(kernel = 'rbf', random_state = 42)
    clf.fit(X_train, y_train)
    y_predict = clf.predict(X_test)
    cm = confusion_matrix(y_test, y_predict)
    print(cm)
    print("Accuracy score of Arousal ")
    print(accuracy_score(y_test, y_predict)*100)
    
    
    ####################################################################
    
    file_x = 'data/features_sampled.dat'
    file_y = 'data/label_class_2.dat'
    
    X = numpy.genfromtxt(file_x, delimiter=' ')
    y = numpy.genfromtxt(file_y, delimiter=' ')
    
    # Split the data into training/testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
    
    # Feature Scaling
    from sklearn.preprocessing import StandardScaler
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    	
    # SVM Classifier
    clf = SVC(kernel = 'rbf', random_state = 42)
    clf.fit(X_train, y_train)
    y_predict = clf.predict(X_test)
    cm = confusion_matrix(y_test, y_predict)
    print(cm)
    print("Accuracy score of Dominance ")
    print(accuracy_score(y_test, y_predict)*100)
    
    ######################################################################3
    
    file_x = 'data/features_sampled.dat'
    file_y = 'data/label_class_3.dat'
    
    X = numpy.genfromtxt(file_x, delimiter=' ')
    y = numpy.genfromtxt(file_y, delimiter=' ')
    
    # Split the data into training/testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
    
    # Feature Scaling
    from sklearn.preprocessing import StandardScaler
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    	
    # SVM Classifier
    clf = SVC(kernel = 'rbf', random_state = 42)
    clf.fit(X_train, y_train)
    y_predict = clf.predict(X_test)
    cm = confusion_matrix(y_test, y_predict)
    print(cm)
    print("Accuracy score of liking  ")
    print(accuracy_score(y_test, y_predict)*100)
```

```
emo['New_valence'] = emo['Valence_mean']/emo['Valence_std']
emo['New_arousal'] = emo['Arousal_mean']/emo['Arousal_std']
emo['New_dominance'] = emo['Dominance_mean']/emo['Dominance_std']
emo
```

**A 2D emotional model** within the Valence-Arousal space that had 4 quadrants i.e. Low Valence Low Arousal (LALV), High Valence Low Arousal (HVLA), Low Valence High Arousal (LVHA), and High Valence High Arousal (HVHA). They established high positive correlations between liking and valence, and between dominance and valence

```
#fundamental emotion
conditions = [
    (emo['Valence_mean'] <= 5)& (emo['Arousal_mean'] <= 5) , #lalv
    (emo['Valence_mean'] <= 5 ) & (emo['Arousal_mean'] >= 5) ,  # halv
    (emo['Valence_mean'] >= 5) & (emo['Arousal_mean'] <= 5),  # lahv
    (emo['Valence_mean'] >= 5 )& (emo['Arousal_mean']>= 5)    # hahv
    ]

values = ['Sad', 'scared', 'calm', 'happy']


emo['Emotion_state'] = np.select(conditions, values)

emo
```

**A 3D emotional model** by introducing an additional parameter Dominance thus, giving rise to 8 different emotional states within the Valence-Arousal-Dominance space. Our approach of preprocessing the Valence, Arousal and Dominance values by using the established Mean of values for classifying the individual classes and incorporating a third dimension. The Average values obtained of all the 3 parameters namely valence, arousal &dominance for set of 32 participants.
Thus, each participant is provided with an emotional state label that depicts the overall emotional state of the participant while performing the experiment.
~Emotion Classification Using EEG Signals.

```
conditions = [
    (emo['Valence_mean'] <= 5)& (emo['Arousal_mean'] <= 5) & (emo['Dominance_mean']<=5), #lalvld
    (emo['Valence_mean'] <= 5 ) &( emo['Arousal_mean'] >= 5) & (emo['Dominance_mean']<=5), # halvld
    (emo['Valence_mean'] >= 5) & (emo['Arousal_mean'] <= 5)&  (emo['Dominance_mean']<=5), # lahvld
    (emo['Valence_mean'] >= 5 )& (emo['Arousal_mean'] >= 5) & (emo['Dominance_mean']<=5),   # hahvld
    (emo['Valence_mean'] <= 5)& (emo['Arousal_mean'] <= 5) &  (emo['Dominance_mean']>=5), #lalvhd
    (emo['Valence_mean'] <= 5 ) &( emo['Arousal_mean'] >= 5) &  (emo['Dominance_mean']>=5), # halvhd
    (emo['Valence_mean'] >= 5) & (emo['Arousal_mean'] <= 5)& (emo['Dominance_mean']>=5), # lahvhd
    (emo['Valence_mean'] >= 5 )& (emo['Arousal_mean'] >= 5)&  (emo['Dominance_mean']>=5)    # hahvhd
    ]
 
values = ['Bored', 'Disgust', 'Peaceful', 'Surprise' , 'Sad' , 'Nervous', 'Relaxed', 'Excited']
 
 
emo['Complex_Emotion_state'] = np.select(conditions, values)
 
emo
```
